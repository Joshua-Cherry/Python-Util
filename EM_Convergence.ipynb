```python

##############################################################################################################################
# I wrote this program to help myself solve some homework questions related to Expecation Maximization
##############################################################################################################################

import sys
import os
from time import time
%matplotlib inline
from urllib.request import urlopen
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import numpy as np
import math




mu1 = -2
mu2 = 2
x1 = -1
x2 = 1
```

```python
# Exponent
n = 2

###############################################################################################################################################
# This calculates the probability of mu1 with respect to the current datapoint x. When calculating the value of mu2 just pass mu2 as mu1 
# or we can subtract mu1 from 1 and et mu2 to whatever value that is.
###############################################################################################################################################
def Prob(mu1, mu2, x):
  sol = math.exp(-((x-mu1)**n)/2)/(math.exp(-((x-mu1)**n)/2) + math.exp(-((x-mu2)**n)/2))
  return sol

##############################################################################################################################
# This calculates the new mean with respect to the two probablities, p1 and p2, and the two datapoints, x1 and x2.
##############################################################################################################################
def em(p1, p2, x1, x2):
  sol = ((p1 * x1) + (p2 * x2))/(p1 + p2)
  return sol
```

```python

##############################################################################################################################
# This value is just calculating the new mean for the first iteration of Expecation maximization
##############################################################################################################################

p1 = Prob(mu1, mu2, x1)
p2 = 1-p1
sol = em(p1, p2, x1, x2)
print(sol)


p_1 = Prob(mu1, mu2, x2)
p_2 = 1-p_1

sol = em(p_1, p_2, x1, x2)
print(sol)
```

```python

##############################################################################################################################
# Running expectation maximization for 1000 iterations. 
##############################################################################################################################
for i in range(1000):
  tmp_mu = mu1
  p1 = Prob(mu1, mu2, x1)
  p2 = 1-p1

  mu1 = em(p1, p2, x1, x2)
  print(mu1)


  p_1 = Prob(mu1, mu2, x2)
  p_2 = 1-p_1

  mu2 = em(p_1, p_2, x1, x2)
  print(mu2)
```
